```python
import pandas as pd
import numpy as np

clinical = pd.read_csv("data/raw/clinical.tsv", sep="\t", low_memory=False)
follow_up = pd.read_csv("data/raw/follow_up.tsv", sep="\t", low_memory=False)
pathology = pd.read_csv("data/raw/pathology_detail.tsv", sep="\t", low_memory=False)

# Fix GDC's missing value marker
for df in [clinical, follow_up, pathology]:
    df.replace("'--", np.nan, inplace=True)

print(f"Patients: {clinical['cases.case_id'].nunique()}")
print(f"Deaths: {(clinical['demographic.vital_status'] == 'Dead').sum()}")
print(f"Alive: {(clinical['demographic.vital_status'] == 'Alive').sum()}")
```
